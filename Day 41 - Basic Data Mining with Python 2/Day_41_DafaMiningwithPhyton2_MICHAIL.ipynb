```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.set_style("whitegrid")
import matplotlib.pyplot as plt
%matplotlib inline

import plotly.express as px

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('sample_data/messy_daily_website_visitors.csv')

df.rename(columns = {
    'Day.Of.Week' : 'day_of_week',
    'Page.Loads' : 'page_loads',
    'Unique.Visits' : 'unique_visits',
    'First.Time.Visits' : 'first_visits',
    'Returning.Visits' : 'returning_visits'
}, inplace = True)

df = df.replace(',', '', regex=True)

df['page_loads'] = df['page_loads'].fillna(0).astype(int)
df['unique_visits'] = df['unique_visits'].fillna(0).astype(int)
df['first_visits'] = df['first_visits'].fillna(0).astype(int)
df['returning_visits'] = df['returning_visits'].fillna(0).astype(int)

df
```

```python
# a. Berapa total returning visits dan nilai rata-rata dari unique visits setelah data dibersihkan?
# Menghitung total returning visits
total_returning_visits = df['returning_visits'].sum()

# Menghitung rata-rata dari unique visits
average_unique_visits = df['unique_visits'].mean()

total_returning_visits, average_unique_visits
```

```python
# b. Menghitung rata-rata jumlah page loads untuk setiap unique visits
average_page_loads_per_unique_visit = df['page_loads'].sum() / df['unique_visits'].sum()

average_page_loads_per_unique_visit
```

```python
# c. Menghitung jumlah page loads per hari (berdasarkan 'day_of_week')
page_loads_per_day = df.groupby('day_of_week')['page_loads'].sum()

page_loads_per_day
```
